```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("prob_target_study.csv")
```

ref: https://www.evidentlyai.com/classification-metrics/classification-threshold

```python
df["Actual_target"] = df["Actual_target"].astype("str")
df["Actual_target"] = df["Actual_target"].str.replace("1.0",  "Positive")
df["Actual_target"] = df["Actual_target"].str.replace("0.0",  "Negative")
```

```python
sns.histplot(data = df, x="Probability", hue="Actual_target", bins=25)
plt.title("Distribution of the probability assigned by the model")
plt.show()
```

```python
positives = df[df["Actual_target"]=="Positive"]
negatives = df[df["Actual_target"]=="Negative"]
```

```python

mean_positives = np.mean(positives["Probability"])
std_positives = np.std(positives["Probability"])
median_positives = np.median(positives["Probability"])

print("Avg. probability for the positive class: {}".format(mean_positives))
print("Std. probability for the positive class: {}".format(std_positives))
print("Median probability for the positive class: {}".format(median_positives))
```

```python
mean_negatives = np.mean(negatives["Probability"])
std_negatives = np.std(negatives["Probability"])
median_negatives = np.median(negatives["Probability"])

print("Avg. probability for the positive class: {}".format(mean_negatives))
print("Std. probability for the positive class: {}".format(std_negatives))
print("Median probability for the positive class: {}".format(median_negatives))
```
